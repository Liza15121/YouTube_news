```python
def get_channel_stats (youtube, channel_ids):
    all_data = []
    request = youtube.channels().list(
        part='snippet,contentDetails,topicDetails,statistics',
        id=','.join(channel_ids)
    )
    response = request.execute()
    
    for i in range(len(response['items'])):
        data = {'channel_name': response['items'][i]['snippet']['title'],
                'channel_id': response['items'][i]['id'],
                'subscribers': response['items'][i]['statistics']['subscriberCount'],
                'views': response['items'][i]['statistics']['viewCount'],
                'total_videos': response['items'][i]['statistics']['videoCount'],
                'description': response['items'][i]['snippet']['description'],
                'playlist_id': response['items'][i]['contentDetails']['relatedPlaylists']['uploads'],
                'country': response['items'][i]['snippet'].get('country', 'NA'),
                'topicDetails': response['items'][i].get('topicDetails', {}),
                'publishedAt': pd.to_datetime(response['items'][i]['snippet']['publishedAt']).tz_convert('Asia/Kolkata')
                
                }
        all_data.append(data)

        
    return all_data
```

```python
def get_video_ids_from_playlists(youtube, playlist_ids, per_playlist_limit=50):
    all_video_ids = []

    for playlist_id in playlist_ids:
        video_ids = []
        request = youtube.playlistItems().list(
            part='contentDetails',
            playlistId=playlist_id,
            maxResults=min(50, per_playlist_limit)
        )
        response = request.execute()

        for item in response['items']:
            video_ids.append(item['contentDetails']['videoId'])

        next_page_token = response.get('nextPageToken')
        while next_page_token and len(video_ids) < per_playlist_limit:
            request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=playlist_id,
                maxResults=min(50, per_playlist_limit - len(video_ids)),
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                video_ids.append(item['contentDetails']['videoId'])

            next_page_token = response.get('nextPageToken')

        all_video_ids.extend(video_ids)

    return all_video_ids
```

```python
video_ids = get_video_ids_from_playlists(youtube, playlist_ids, per_playlist_limit=50)
```

# Function to get videos details

```python
def get_video_details(youtube, video_ids):
    all_video_data = []

    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part='snippet,statistics,contentDetails',
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute()

        for item in response['items']:
            video_data = {
                'video_id': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'viewCount': item['statistics'].get('viewCount', 0),
                'likeCount': item['statistics'].get('likeCount', 0),
                'dislikeCount': item['statistics'].get('dislikeCount', 0),
                'commentCount': item['statistics'].get('commentCount', 0),
                'duration': item['contentDetails']['duration'],
                'publishedAt': pd.to_datetime(item['snippet']['publishedAt']).tz_convert('Asia/Kolkata')

            }
            all_video_data.append(video_data)

    return all_video_data
```

```python
video_details = get_video_details(youtube, video_ids)
```

```python
df_video_data = pd.DataFrame(video_details)
```
